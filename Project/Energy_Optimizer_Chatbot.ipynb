```python
import google.generativeai as genai

# Setup Gemini API
genai.configure(api_key="AIzaSyDZUqQThj5vvjDXUdEMW4-NJChpyZ6sPTs")
model = genai.GenerativeModel("gemini-1.5-flash-latest")
chat = model.start_chat(history=[])

# Intro
def intro():
    print("\n⚡ Welcome to Energy Optimizer Bot (SDG 7) ⚡")
    print("Helping households and businesses save energy 🌱")
    print("Ask me things like:\n- How can I reduce my electricity bill?\n- Recommend efficient air conditioners\n- Should I install solar panels?\nType 'exit' to quit.\n")

# Chat loop
def chat_loop():
    intro()
    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Bot: Stay energy efficient! 🌞 Goodbye!")
            break

        try:
            prompt = (
                f"You are a smart energy assistant designed to support SDG 7. "
                f"Provide helpful, clear, and practical answers to help users reduce energy consumption, "
                f"suggest efficient appliances, and educate them on renewable energy options.\n\nUser: {user_input}"
            )

            response = chat.send_message(prompt)
            print("Bot:", response.text)
        except Exception as e:
            print("Error:", e)

# Run the bot
if __name__ == "__main__":
    chat_loop()
```
